```python
import pandas as pd
import matplotlib.pyplot as plt
#누적변수 설정
a = [0,0] #위치
fb = 0 #웅덩이 횟수
fg = 0 #잔디 기분

#기본 밑그림
plt.figure(figsize = (10,10))
plt.style.use('ggplot') #격자무늬
plt.xlim(0,20) #x축범위제한
plt.ylim(0,20)
plt.xticks(range(0,21)) #x축 눈금
plt.yticks(range(0,21))
plt.scatter(5,15, s = 13200, color = 'b', marker = 'D', alpha = 0.2)
plt.scatter(8,12, s = 6000, color = 'b', marker = 'D', alpha = 0.2)
plt.scatter(2,6, s = 1500, color = 'b', marker = 'D', alpha = 0.2)
plt.scatter(14,6, s = 1500, color = 'b', marker = 'D', alpha = 0.2)
plt.scatter(17,3, s = 6000, color = 'b', marker = 'D', alpha = 0.2)
plt.scatter(2,2, s = 3000, color = 'r', marker = 'o', alpha = 0.2)
plt.scatter(5,5, s = 12000, color = 'r', marker = 'o', alpha = 0.2)
plt.scatter(7,7, s = 3000, color = 'r', marker = 'o', alpha = 0.2)
plt.scatter(8,3, s = 12000, color = 'r', marker = 'o', alpha = 0.2)
plt.scatter(12,12, s = 12000, color = 'r', marker = 'o', alpha = 0.2)
#여기까지 웅덩이와 잔디밭 그리기


#여기부터 프로그래밍파트

#웅덩이와 잔디의 위치
pool1 = [[2,2],[7,7],2]
pool2 = [[5,5],[8,3],[12,12],3]
grass1 = [[2,6],[14,6],2]
grass2 = [[17,3],[8,12],3]
grass3 = [[5,15],4]
pool = [pool1,pool2]
grass = [grass1,grass2,grass3]
#pool = [[2,1],[2,2],[2,3],[1,2][3,2],[]

#입력파트
while True:
  move = input()
  if move in ['w','a','s','d','W','A','S','D']:
    break

#이동 파트
#a = [0,0] # 이건 맨 앞에 반복문 전에 설정해주면 될듯
if move == 'w' or move == 'W':
  a[1] = a[1] + 1
elif move == 'a' or move == 'A':
  a[0] = a[0] - 1
elif move == 's' or move == 'S':
  a[1] = a[1] - 1
elif move == 'd' or move == 'D':
  a[0] = a[0] + 1

#산책코스를 벗어났는지
if a[0] < 0 or a[1] < 0 or a[0] > 20 or a[1] > 20:
  break




#잔디 혹은 웅덩이 밟았는지
for i in grass:
  for j in range(0,len(i)-1):
    if abs(a[0]-i[j][0])+abs(a[1]-i[j][1]) < i[-1]:
      print("'코코가 매우 기뻐합니다!'")
      fg = fg + 10

for i in pool:
  for j in range(0,len(i)-1):
    if abs(a[0]-i[j][0])+abs(a[1]-i[j][1]) < i[-1]:
      print("'코코의 기분이 좋지 않아 보입니다.'")
      fb = fb + 1

#기분 많이나쁜지
if fb = 3:
  break

#반복문 끝나고 최종 출력
print('산책 끝')
print('코코의 기분이',fg,'좋아졌습니다!')



#별 표시
plt.scatter(a[0],a[1], s = 1000, color = 'g', marker = '*', alpha = 0.9)


plt.show
```
